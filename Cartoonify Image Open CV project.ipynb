```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('test.jpg')
```

```python
#BGR to RGB conversion
image=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
#reducing pixelsize
img_small=cv2.pyrDown(image)
```

```python
#Applying Bilateral Filter
i=5
for _ in range(i):
    img_small=cv2.bilateralFilter(img_small, d=9, sigmaColor=9, sigmaSpace=7)
```

```python
img2=cv2.pyrUp(img_small)
```

```python
plt.imshow(img2)
plt.show()
```

```python
img_gray=cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
img_blur=cv2.medianBlur(img_gray,7)
```

```python
img_edge=cv2.adaptiveThreshold(img_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 7, 2)
```

```python
plt.imshow(img_edge)
plt.show()
```

```python
img_edge=cv2.cvtColor(img_edge, cv2.COLOR_GRAY2RGB)
```

```python
array= cv2.bitwise_and(img_edge,img2)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(array)
plt.show()
```
